```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
%matplotlib inline
```

```python
x = np.arange(0,5)
y1 = np.array([71.53, 88.78, 78.53, 100.00, 84.12])
y2 = np.array([61.13, 88.11, 64.58, 98.17, 70.82])
```

```python
plt.figure(figsize=(10,8))
plt.xlabel('Classe', fontsize=15)
plt.ylabel('precisão (em porcento)', fontsize=15)
bar_width = 0.25
plt.xticks(x + bar_width/2,x, fontsize=15)
plt.yticks(fontsize=15)
temp = 0.05

for i in x:
    value = str(y1[i]) + "%"
    plt.text(i - bar_width/2 - temp - 0.1, y1[i]+0.5, value, fontsize=15)
    
for i in x:
    value = str(y2[i]) + "%"
    plt.text(i + bar_width/2 + temp, y2[i]+0.5, value, fontsize=15)

plt.bar(x - temp,y1, color='#cccccc', width=bar_width, label='sem descritores')
plt.bar(x + bar_width + temp,y2, color='#656565', width=bar_width, label='com descritores')

plt.legend(fontsize=15, shadow=True, ncol=2, loc='upper center', bbox_to_anchor=(0.5, -0.1))
plt.tight_layout()
plt.savefig('gráfico1_cinza.png')
```
