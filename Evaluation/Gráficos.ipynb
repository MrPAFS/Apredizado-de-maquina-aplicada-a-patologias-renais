```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,5)
y1 = np.array([67.58, 88.45, 71.99, 100.00, 80.67])
y2 = np.array([60.55, 86.29, 58.95, 95.33, 75.06])
```

```python
plt.figure(figsize=(12,10))
plt.xlabel('Classe')
plt.ylabel('Precisão (em porcento)')

for i in x:
    value = str(y1[i]) + "%"
    plt.text(i-0.25, y1[i]+0.5, value, fontsize=12)

plt.bar(x,y1, color='#00a0a0', width=0.5)

plt.tight_layout()
plt.savefig('gráfico1.png')
```

```python
plt.figure(figsize=(12,10))
plt.xlabel('Classe')
plt.ylabel('Precisão (em porcento)')

for i in x:
    value = str(y2[i]) + "%"
    plt.text(i-0.25, y2[i]+0.5, value, fontsize=12)

plt.bar(x,y2, color='#00a0a0', width=0.5)

plt.tight_layout()
plt.savefig('gráfico2.png')
```
