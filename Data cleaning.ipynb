```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
for (column, data) in database.iteritems():
    print(column + ": ")
    
    unique = data.unique()
    
#     i = 0
#     while i < len(unique):
    for i in range(len(unique)):
#         j = i + 1
#         while j < len(unique):
        for j in range(i+1, len(unique)):
            rate = similar(unique[i], unique[j])
            if rate >= 0.8:
                print("[" + str(unique[i]) + ", " + str(unique[j]) + "] = " + str(rate))
#             j += 1
#         i += 1
    print("\n")
```

```python
def correct_Imunodepositos(reg):
    if reg == 'IgG IgA IgM C3 fibrinogênio kappa lambda':
        return 'IgG IgA IgM C3 Fibrinogênio kappa lambda'
    else:
        return reg
    
def correct_Clinica(reg):
    if reg == 'proteinúria hematúria':
        return 'proteinúria e hematúria'
    elif reg == 'síndrome nefrítico-nefrótica':
        return 'síndrome nefrítica-nefrótica'
    else:
        return reg
def correct_Diagnostico(reg):
    if reg == 'GLOMERULONEFRITE LÚPICA DIFUSA E GLOBAL- CLASSE IV':
        return 'GLOMERULONEFRITE LÚPICA DIFUSA E GLOBAL - CLASSE IV'
    elif reg == 'Gloemrulonefrite crescêntica':
        return 'Glomerulonefrite crescêntica'
    elif reg == 'GLOMERULONEFRITE LÚPICA SEGMENTAR E FOCAL – CLASSE III':
        return 'GLOMERULONEFRITE LÚPICA SEGMENTAR E FOCAL - CLASSE III'
    else:
        return reg
```

```python
database['Imunodepósitos'] = database['Imunodepósitos'].apply(lambda reg: correct_Imunodepositos(reg))
database['Clínica'] = database['Clínica'].apply(lambda reg: correct_Clinica(reg))
database['Diagnóstico'] = database['Diagnóstico'].apply(lambda reg: correct_Diagnostico(reg))
```

```python
database.to_csv('base-100-limpa+.csv')
```
